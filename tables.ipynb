```python
from astropy.table import Table
import numpy as np
from astropy.io import ascii
from astropy import units as u
from astropy import constants as cons
```

#### ALMA-IMF

```python
louvet24 = Table.read("cmf_zoo/data/download/louvet+2024.fit")
```

```python
gamma_mask = louvet24['Sint1_3mm'] / (louvet24['Sint3mm']+1e-9) >= 2
region_mask = louvet24['Cluster'] == 'W51-E'
mask = (~region_mask) & (gamma_mask)
louvet24_mask = louvet24[mask]
output_cores = louvet24_mask['CoreName', 'Mass', 'e_Mass', 'T', 'Size']

output_cores.rename_columns(
    names=['CoreName', 'Mass', 'e_Mass', 'T', 'Size', ],
    new_names=["ID", "Mcore", "err_Mcore", "Tcore", "Size"],
)
```

```python
output_cores.write(
    "cmf_zoo/data/catalogues/almaimf.csv",
    formats={"Mcore": "%.2f", "err_Mcore": "%.2f", "Tcore": "%.1f", "Size": "%02d"},
    overwrite=True,
)
```

#### ASHES

```python
morii23 = Table.read("cmf_zoo/data/download/ASHES_core_share_small.csv")
clump_mask = [each not in ['G010.99', 'G014.49', 'G018.80', 'G024.01', 'G028.56'] for each in morii23['ClumpName']]
```

```python
output_cores = morii23[clump_mask]
output_cores = output_cores['col0', 'Mass', 'T', 'radius_au']
output_cores.add_column(output_cores['Mass'] * 0.5, index=2, name='e_Mcore')
output_cores.rename_columns(
    names=['col0', 'Mass', 'e_Mcore', 'T', 'radius_au', ],
    new_names=["ID", "Mcore", "err_Mcore", "Tcore", "Size"],
)
```

```python
output_cores.write(
    "cmf_zoo/data/catalogues/ashes.csv",
    formats={"Mcore": "%.2f", "err_Mcore": "%.2f", "Tcore": "%.1f", "Size": "%02d"},
    overwrite=True,
)
```

#### LMC

```python
with open("cmf_zoo/data/download/traficante+2026.txt", "r") as f:
    header_line = f.readline()

names = [
    col.strip()
    for col in header_line.strip().split("|")
    if col.strip()
]

traficante26 = Table.read(
    "cmf_zoo/data/download/traficante+2026.txt",
    format="ascii.basic",
    names=names,
    data_start=4,
    guess=False,
)
```

```python
output_cores = traficante26['CORE_ID', 'MASS', 'RADIUS']
output_cores.add_column(traficante26['MASS'] * 0.5, index=2, name='err_Mcore')
output_cores.add_column(np.full(71, 36.7), index=3, name='Tcore')
output_cores.rename_columns(
    names=['CORE_ID', 'MASS', 'err_Mcore', 'Tcore', 'RADIUS', ],
    new_names=["ID", "Mcore", "err_Mcore", "Tcore", "Size"],
)
```

```python
output_cores.write(
    "cmf_zoo/data/catalogues/lmc.csv",
    formats={"Mcore": "%.2f", "err_Mcore": "%.2f", "Tcore": "%.1f", "Size": "%02d"},
    overwrite=True,
)
```

#### Aquila

```python
Konyves15 = Table.read("cmf_zoo/data/download/Konyves+2015.fit")
```

```python
output_cores = Konyves15[Konyves15["Coretype"] != 'protostellar']
output_cores = output_cores["Seq", "Mcore", "e_Mcore", "Tdust", "Rd"]
output_cores.replace_column("Rd", col=output_cores["Rd"].to(u.au))
output_cores.rename_columns(
    names=['Seq', 'Mcore', 'e_Mcore', 'Tdust', 'Rd', ],
    new_names=["ID", "Mcore", "err_Mcore", "Tcore", "Size"],
)
```

#### Quarks

```python
jiao26 = Table.read("cmf_zoo/data/download/apj_table3_full.tex", format='latex')
```

```python
coreID = [f"{each['Source']}_{each['ID']}" for each in jiao26]
err_Mcore = jiao26['Mass'] * 0.5
output_cores = jiao26['Mass', 'Tdust', 'Rdec']
output_cores.add_columns(cols=[coreID, err_Mcore], names=['ID', 'err_Mass'], indexes=[0, 1])

output_cores.rename_columns(
    names=['ID', 'Mass', 'err_Mass', 'Tdust', 'Rdec', ],
    new_names=["ID", "Mcore", "err_Mcore", "Tcore", "Size"],
)
```

```python
output_cores.write(
    "cmf_zoo/data/catalogues/quarks.csv",
    formats={"Mcore": "%.2f", "err_Mcore": "%.2f", "Tcore": "%.1f", "Size": "%02d"},
    overwrite=True,
)
```

### Central molecular zone

```python
from astropy.io import ascii

zhang2025 = ascii.read("cmf_zoo/data/download/apjad9f28t1_mrt.txt")
```

```python
coreID = [f"{each["Cloud"]}_{each["Core"]}" for each in zhang2025]
err_Mcore = zhang2025['Mcore'] * 0.5

output_cores = zhang2025['Mcore', 'T', 'R']
output_cores.add_columns(cols=[coreID, err_Mcore], names=['ID', 'err_Mass'], indexes=[0, 1])

output_cores.rename_columns(
    names=['ID', 'Mcore', 'err_Mass', 'T', 'R', ],
    new_names=["ID", "Mcore", "err_Mcore", "Tcore", "Size"],
)
```

```python
output_cores.write(
    "cmf_zoo/data/catalogues/cmz.csv",
    formats={"Mcore": "%.2f", "err_Mcore": "%.2f", "Tcore": "%.1f", "Size": "%02d"},
    overwrite=True,
)
```
